# Doctor Recommender System

```python
import pandas as pd
import numpy as np
import warnings
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

from geopy.geocoders import Nominatim
```

```python
doctors = pd.read_csv(r"C:\Users\snehal\PycharmProjects\PetoCare\recommenderSystem\PetoCareDoc.csv")
doctors
```

```python
doctors = doctors.dropna(axis=1)
doctors
```

```python
docLocation = doctors.iloc[:,-2:]
docLocation
```

```python
kmeans = KMeans(5)
kmeans.fit(docLocation)
kmeans
```

```python
identifiedCluster = kmeans.fit_predict(docLocation)
identifiedCluster = list(identifiedCluster)
```

```python
identifiedCluster
```

```python
doctors["loc_cluster"] = identifiedCluster
doctors
```

```python
input_area = input("Enter a area name: ")
areaList = list(input_area.split(' '))
areaList = [area.capitalize() for area in areaList]
input_area = " ".join(areaList)
print("input_area =",input_area)
cluster = doctors.loc[doctors['Area'] == input_area, 'loc_cluster']
print("cluster size", len(cluster))
if len(cluster) > 0:
    cluster = cluster.iloc[0]
    print("cluster = ", cluster)
    areas = doctors.loc[doctors['loc_cluster'] == cluster, 'Area']
    #     print("areas =",areas)
    for c in range(len(areas)):
        print(areas.iloc[c])
        #         if areas.iloc[c] == input_area:
        #             continue
        #         else:
        #             print(areas.iloc[c])
```
