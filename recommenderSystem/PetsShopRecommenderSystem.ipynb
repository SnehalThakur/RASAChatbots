# Pets Shop Recommender System

```python
import pandas as pd
import numpy as np
import warnings
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

from geopy.geocoders import Nominatim
```

```python
petShop = pd.read_csv(r"C:\Users\snehal\PycharmProjects\PetoCare\recommenderSystem\petshopss.csv")
petShop
```

```python
petShop = petShop.dropna(axis=1)
petShop
```

```python
shopLocation = petShop.iloc[:,-2:]
shopLocation
```

```python
kmeans = KMeans(5)
kmeans.fit(shopLocation)
kmeans
```

```python
identifiedCluster = kmeans.fit_predict(shopLocation)
identifiedCluster = list(identifiedCluster)
```

```python
identifiedCluster
```

```python
petShop["loc_cluster"] = identifiedCluster
petShop
```

```python
input_area = input("Enter a area name: ")
areaList = list(input_area.split(' '))
areaList = [area.capitalize() for area in areaList]
input_area = " ".join(areaList)
print("input_area =",input_area)
cluster = petShop.loc[petShop['Area'] == input_area, 'loc_cluster']
print("cluster size", len(cluster))
if len(cluster) > 0:
    cluster = cluster.iloc[0]
    print("cluster = ", cluster)
    areas = petShop.loc[petShop['loc_cluster'] == cluster, 'Area']
    #     print("areas =",areas)
    for c in range(len(areas)):
        print(areas.iloc[c])
        #         if areas.iloc[c] == input_area:
        #             continue
        #         else:
        #             print(areas.iloc[c])
```
